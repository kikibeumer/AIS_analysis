```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
ais = pd.read_csv("Dataset_30July.csv",sep=",")
ais.head()
```

ROT

```python
ais5 = ais[(ais['Message Type'] == 1) | (ais['Message Type'] == 2) | (ais['Message Type'] == 3)]

ais5
```

```python
# Ensure that datatype is float
ais.loc[ais['ROT'] == 'Not turning', 'ROT'] = 0
ais.loc[ais['ROT'] == 'No turn info available', 'ROT'] = 800
ais.loc[ais['ROT'] == float('nan'), 'ROT'] = 800
ais.loc[ais['ROT'] == 'NaN', 'ROT'] = 800

ais['ROT'] = pd.to_numeric(ais['ROT'], errors='coerce')

# Replace any remaining NaN values with 800
ais['ROT'].fillna(800, inplace=True)

# Ensure the datatype is float
ais['ROT'] = ais['ROT'].astype(float)
ais["ROT"]

```

```python
ais.dtypes
```

```python
ais_filtered = ais[ais['ROT'] != 800.0]
ais_filtered
```

SOG

```python
ais['SOG'] = pd.to_numeric(ais['SOG'], errors='coerce')

# Replace any remaining NaN values with 800
ais['SOG'].fillna(102.3, inplace=True)

# Ensure the datatype is float
ais['SOG'] = ais['SOG'].astype(float)
ais["SOG"]
```

```python
ais_sog = ais[ais['SOG'] != 102.3]
ais_sog

plt.hist(ais_sog["SOG"], range=(0,25), bins=100)
plt.show() 
```

```python
sns.set_theme(style="darkgrid")
sns.histplot(data=ais_sog, x="SOG", kde=True)
plt.xlim(0, 25)

plt.show()
```

A, B, C, D

```python
ais_length = ais.dropna(subset='A')
ais_length = ais_length.loc[ais_length['A'] != 0]
```

```python
plt.figure(figsize=(12,8))

sns.histplot(ais_length['A'], bins=30, color='blue', label='A', kde=True)
sns.histplot(ais_length['B'], bins=30, color='green', label='B', kde=True)
sns.histplot(ais_length['C'], bins=30, color='red', label='C', kde=True)
sns.histplot(ais_length['D'], bins=30, color='purple', label='D', kde=True)

plt.legend()
plt.title('Histograms of vessel length')
plt.xlabel('Length')
plt.ylabel('Frequency')

plt.show()
```
