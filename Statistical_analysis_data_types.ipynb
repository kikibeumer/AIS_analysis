Ship type

```python
# Create the count table
ship_type_table = ais['Ship type'].value_counts(dropna=True).reset_index()

# Display the count table
print(ship_type_table)
```

```python
# Define the mapping of conditions to their respective values
conditions = [
    (ais['Ship type'].str.contains('Cargo', case=False, na=False), 'Cargo'),
    (ais['Ship type'].str.contains('Tanker', case=False, na=False), 'Tanker'),
    (ais['Ship type'].str.contains('Passenger', case=False, na=False), 'Passenger'),
    (ais['Ship type'].str.contains('High Speed Craft', case=False, na=False), 'High Speed Craft')
]

# Apply the conditions to update the 'ship_type' column
for condition, value in conditions:
    ais.loc[condition, 'Ship type'] = value

# Set anything else to 'other'
ais.loc[~ais['Ship type'].isin(['Cargo', 'Tanker', 'Passenger', 'High Speed Craft']), 'Ship type'] = 'other'

ais
```

```python
ship_type_table.columns = ['Ship type', 'Count']

# Plot the count table using seaborn
sns.barplot(x="Ship type", y="Count", data=ship_type_table)

# Add labels and title
plt.xlabel('Ship type')
plt.ylabel('Count')
plt.title('Ship type Counts')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
